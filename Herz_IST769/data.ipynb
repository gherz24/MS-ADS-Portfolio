```python
### create 'table_name', 'column family 1', 'column family 2', 'column family 3'
        ### create table

### put 'table_name', 'row_key', 'column family:column name', 'value'
        ### input data through shell

teams = ["NewYork Mets", 
         "Chicago Cubs", 
         "StLouis Cardinals", 
         "Boston RedSox", 
         "Seattle Mariners"]

hand = ["Right","Left"]

import statistics
import random
import numpy.random as npr

print("""Copy & Paste Code into Hbase Shell 
      \ncreate 'baseball_data', 'player_info', 'player_stats', 'team_info', 'team_stats'""")

x = 0
for team in teams:
    city = str.split(team, " ")[0]
    team_name = str.split(team, " ")[1]
    wins = random.randrange(50, 110)
    losses = 162 - wins
    for player in range(1):
        x += 1
        player_number = random.randrange(99)
        country = random.choice(["USA", "Canada", "Mexico", "Cuba", "Puerto Rico", "Japan", "Korea"])
        bat = random.choice(hand)
        throw = random.choice(hand)
        position = random.choice(["C", "1B", "2B", "3B", "SS", "OF"])
        gp = random.randrange(20,162)
        ab = int(round(gp * npr.normal(3, .2),0))
        avg = round(npr.normal(.25,.02),3)
        hits = int(round(ab*avg, 0))
        hr = int(round(hits*npr.normal(1/7,.05),0))
        print(f"put 'baseball_data', '{x}', 'player_info:player_number', {player_number}")
        print(f"put 'baseball_data', '{x}', 'player_info:position', '{position}'")
        print(f"put 'baseball_data', '{x}', 'player_info:country', '{country}'")
        print(f"put 'baseball_data', '{x}', 'player_info:bat', '{bat}'")
        print(f"put 'baseball_data', '{x}', 'player_info:throw', '{throw}'")
        print(f"put 'baseball_data', '{x}', 'player_stats:games_played', {gp}")
        print(f"put 'baseball_data', '{x}', 'player_stats:ab', {ab}")
        print(f"put 'baseball_data', '{x}', 'player_stats:hits', {hits}")
        print(f"put 'baseball_data', '{x}', 'player_stats:avg', {avg}")
        print(f"put 'baseball_data', '{x}', 'player_stats:homeruns', {hr}")
        print(f"put 'baseball_data', '{x}', 'team_info:city', '{city}'")
        print(f"put 'baseball_data', '{x}', 'team_info:name', '{team_name}'")
        print(f"put 'baseball_data', '{x}', 'team_stats:wins', {wins}")
        print(f"put 'baseball_data', '{x}', 'team_stats:losses', {losses}")
```
